```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('train.csv')

#
```

We observe a quadratic relationship between temperature and differenced values while accounting for month. Intuitively, the energy usage increases for the tail ends of the temperature spectrum and this aligns with the hot vs cold months. 

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x=train['Temperature'], 
                y=train['tv delta'],
               hue=train['Month'],
                legend='full', palette=sns.cubehelix_palette(12))

plt.title("Relationship Between Temperature and Differenced Values")
plt.xlabel("Temperature")
plt.ylabel("Differenced Values")
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x=train['tv delta'], 
                y=train['holiday'],
                palette=sns.cubehelix_palette(2))

plt.title("Relationship Between Temperature and Differenced Values")
plt.xlabel("Differenced Values")
plt.ylabel("Weekday")
```

The weekends (days 5 and 6) have a shorter range. We could attempt using a different model on these days.

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x=train['tv delta'], 
                y=train['Weekday'],
                palette=sns.cubehelix_palette(7))

plt.title("Relationship Between Weekday and Differenced Values")
plt.xlabel("Differenced Values")
plt.ylabel("Weekday")
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x=train['tv delta'], 
                y=train['year'])

plt.title("Relationship Between Temperature and Differenced Values")
plt.xlabel("Differenced Values")
plt.ylabel("Year")
```
